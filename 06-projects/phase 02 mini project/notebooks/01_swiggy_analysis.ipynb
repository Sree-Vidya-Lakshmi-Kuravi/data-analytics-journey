```python
#loading the datasets
import pandas as pd
import numpy as np
```

```python
#loading the excel worksheets/workbooks
file_path = r"E:\Vidya Career\IT JOB\Data Analyst Path\Projects for Data Analysis\Mini PRoject 1\swiggy_dataset.xlsx"

# Load all sheets
orders = pd.read_excel(file_path, sheet_name="orders")
customers = pd.read_excel(file_path, sheet_name="customers")
restaurants = pd.read_excel(file_path, sheet_name="restaurants")
delivery_partners = pd.read_excel(file_path, sheet_name="delivery partners")
```

```python
#converting the date and time columns to its data type
orders['order_time'] = pd.to_datetime(orders['order_time'])
orders['delivery_time'] = pd.to_datetime(orders['delivery_time'])

orders.columns
```

```python
#to see how many cancelled orders
orders['order_status'].value_counts()
```

```python
#successful orders copy
orders_success = orders[orders['order_status'] == 'Successful'].copy()
```

```python
#Total order value
orders_success['total_order_value'] = (orders_success['price'] * orders_success['quantity'])
```

```python
#renaming the column names to make easy to analyze
orders.rename(columns = 
    { 'delivery time (in mins)': 'delivery_time_mins',
      'expected delivery time (in mins)': 'expected_delivery_time_mins'},
    inplace = True)

orders.columns       
```

```python
#renaming the column names to make easy to analyze
orders_success.rename(columns = 
    { 'delivery time (in mins)': 'delivery_time_mins',
      'expected delivery time (in mins)': 'expected_delivery_time_mins'},
    inplace = True)

orders_success.columns
```

```python
#Delivery delay
orders_success['delivery_delay'] = (
    orders_success['delivery_time_mins'] -
    orders_success['expected_delivery_time_mins']
)
```

```python
orders_success[['total_order_value', 'delivery_delay']].head()
```

### Markdown
 Helps convert raw data into business metrics.

```python
#Revenue by Restaurant
revenue_by_restaurant = (
    orders_success
    .groupby('restaurant_id')['total_order_value']
    .sum()
    .reset_index()
    .sort_values(by='total_order_value', ascending=False)
)
```

```python
#merging the restaurant names
revenue_by_restaurant = revenue_by_restaurant.merge(
    restaurants,
    on='restaurant_id',
    how='left'
)
```

```python
revenue_by_restaurant.head(10)
```

### Insight
 Top restaurants contribute a share of total revenue, indicating strong brand or customer preference concentration.

```python
#Top Ordered Dishes
top_dishes = (
    orders_success
    .groupby('dish_ordered')
    .size()
    .reset_index(name='order_count')
    .sort_values(by='order_count', ascending=False)
)
```

```python
top_dishes.head(10)
```

### Insight:
 A small set of dishes dominates customer demand, suggesting opportunities for menu optimization and promotions.

```python
#Top Customers by Revenue
top_customers = (
    orders_success
    .groupby('customer_id')['total_order_value']
    .sum()
    .reset_index()
    .sort_values(by='total_order_value', ascending=False)
)
```

```python
#merging the customer data
top_customers = top_customers.merge(
    customers,
    on='customer_id',
    how='left'
)
```

```python
top_customers.head(10)
```

### Insight:
  High-value customers generate significantly more revenue, indicating potential for loyalty and retention programs.

```python
#Order Success vs Cancelled %
order_status_dist = (
    orders['order_status']
    .value_counts(normalize=True)
    .reset_index()
)
order_status_dist.columns = ['order_status', 'percentage']
order_status_dist['percentage'] = order_status_dist['percentage'] * 100

order_status_dist
```

### Insight:
  Although most orders are successful, cancellations still represent a measurable revenue leakage.

```python
#Calculate average delivery delay per partner
avg_delay_partner = (
    orders_success
    .groupby('delivery_partner_id')['delivery_delay']
    .mean()
    .reset_index()
)
```

```python
avg_delay_partner.head()
```

```python
#Merge with delivery partner table
avg_delay_partner = avg_delay_partner.merge(
    delivery_partners,
    left_on='delivery_partner_id',
    right_on='partner_id',
    how='left'
)
```

```python
avg_delay_partner.head()
```

```python
#Sort for performance insight
#Lower delay = better performance
avg_delay_partner_sorted = avg_delay_partner.sort_values(
    by='delivery_delay'
)

avg_delay_partner_sorted.head(10)
```

### Markdown Insight
- Delivery partners show noticeable differences in average delivery delay.
- Top-performing partners consistently deliver orders faster, highlighting opportunities to replicate best practices and improve overall delivery efficiency.

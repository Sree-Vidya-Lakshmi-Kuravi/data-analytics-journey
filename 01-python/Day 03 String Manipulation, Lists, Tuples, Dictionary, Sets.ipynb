**01. Write a program to get Fibonacci series upto 10 numbers.**

```python
a = 0
b = 1
n = int(input("Enter the number: "))
if n==1:
    print(n)
else:
    for i in range(2,n):
        c = a + b
        a = b
        b = c
        print(c, end = " ")
```

**02. Write a program to check if a number if a number is prime or not.**

```python
num = int(input("Enter the number: "))
if num <= 1:
    print("Not a prime number")

else:
    for i in range(2, num):
        if num%i == 0:
            print("Not a prime number")
            break
        else:
            print("Prime number")
            break
```

**03. Write a program to find a palindrome of integers.**

```python
num = int(input("Enter the number: "))
temp = num
rev = 0

while num>0:
    dig = num%10
    rev = rev*10 + dig
    num = num//10

if rev == temp:
    print("Palindrome")
else:
    print("Not a palindrome")
```

**04. Write a program to create an area calculator.**

```python
print("******AREA CALCULATOR******")

while True:
    print("1: Rectangle \n 2: Square \n 3 : Triangle \n 4 : Circle")

    ch = int(input("Enter number from 1 to 4: "))

    if ch == 1:
        while True:
            l = float(input("Length of rectangle: "))
            b = float(input("Breadth of rectangle: "))
            print("Area of rectangle: ", l * b)
            repeat = input("Want to calculate the area of square again? (y/n)")
            if repeat=='n' or repeat=='N':
                break
                
    
    elif ch == 2:
        while True:
            side = float(input("Enter the side of square: "))
            print("Area of square: ", side ** 2)
            repeat = input("Want to calculate the area of square again? (y/n)")
            if repeat=='n' or repeat=='N':
                break
    
    elif ch == 3:
        while True:
            base = float(input("Enter the base of triangle: "))
            height = float(input("Enter the height of triangle: "))
            print("Area of triangle: ", 0.5 * base * height)
            repeat = input("Want to calculate the area of square again? (y/n)")
            if repeat=='n' or repeat=='N':
                break
    
    elif ch == 4:
        while True:
            radius = float(input("Enter the radius of circle: "))
            print("Area of circle: ", 3.14 * radius * radius)
            repeat = input("Want to calculate the area of square again? (y/n)")
            if repeat=='n' or repeat=='N':
                break
    
    else:
        print("Invalid Choice..Try Again :) ")

    repeat_menu = input("Want to repeat the menu again? (y/n)")
    if repeat_menu == 'n' or repeat_menu == 'N':
        break
```


**Set Functions**

```python
a = {"sambrama","vidya", "pradeep", "vijay", "ram"}

a.add("sravani")  #adds the element anywhere in the set
print(a)

a.pop() #removes any element
print(a)

a = {"sambrama","vidya", "pradeep", "vijay", "ram"}
a.remove("vidya")  #removes given element
print(a)

a.discard("sambrama")
print(a)

b = a.copy() #copies the set
print(b)
```

```python
a = {"vidya", "pradeep", "vijay", "ram"}
b = {"sravani", "sambrama", "lakshmi"}
c = {"vijay", "pradeep"}

#checks if the elements of 2nd set is present in 1st set. Returns True if no same elements.
print(a.isdisjoint(b))
print(a.isdisjoint(c))

#checks whether it is a subset
print(c.issubset(b))
print(c.issubset(a))

#checks whether it is a superset
print(b.issuperset(a))
print(a.issuperset(c))

#unions the sets and removes the duplicates
a.update(c)
print(a)

#clears the entire set
a.clear()
print(a)
```

```python
a = {"vidya", "pradeep", "vijay", "ram"}
b = {"sravani", "sambrama", "lakshmi"}
c = {"vijay", "pradeep"}
 
print(a.union(c)) #combines both sets and removes the duplicate elements

print(a.difference(c)) #removes the common elements and display non-common elements

a.difference_update(c)
print(a)

a = {"vidya", "pradeep", "vijay", "ram"}
b = {"sravani", "sambrama", "lakshmi"}
c = {"vijay", "pradeep"}
print(a.intersection(b))  #Gives the common elements
print(a.intersection(c))

a.intersection_update(c)
print(a)

a = {"vidya", "pradeep", "vijay", "ram"}
b = {"sravani", "sambrama", "lakshmi", "vidya"}
c = {"vijay", "pradeep"}

a.symmetric_difference(b)
print(a)
a.symmetric_difference_update(b)
print(a)
```

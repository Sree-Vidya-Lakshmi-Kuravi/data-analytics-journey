## What is Data Visualization?

Data visualization is the graphical representation of data and information.
It helps in understanding trends, patterns, and insights from data, especially when working with large datasets.

## Matplotlib

Matplotlib is a low-level Python visualization library used for creating static, animated, and interactive plots.

## Bar Plot

```python
import matplotlib.pyplot as plt

x = ["Part1","Part2","Part3","Part4","Part5"]
y = [98,67,88,95,88]

plt.bar(x, y)
plt.xlabel("Parts")
plt.ylabel("Popularity")
plt.title("Popularity of Movie Parts")
plt.show()
```

## Line Plot

```python
import matplotlib.pyplot as plt

x = ["Day1","Day2","Day3","Day4","Day5"]
y1 = [300,420,250,230,400]
y2 = [500,450,300,250,320]

plt.plot(x, y1, label="Week 1")
plt.plot(x, y2, label="Week 2")
plt.legend()
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 10, 50)
y = np.random.randint(10, 100, 50)

plt.scatter(x, y)
plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt

labels = ["Apple","Samsung","Redmi","Nokia"]
sizes = [30,25,20,25]

plt.pie(sizes, labels=labels, autopct="%.1f%%", startangle=90)
plt.show()
```

## Box Plot

```python
import matplotlib.pyplot as plt

data = [25,46,57,25,45,13,29,39,46,99]
plt.boxplot(data)
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt

data = [30,40,68,27,47,59,47,88,44,23]
plt.hist(data, bins=10)
plt.show()
```

## Violin Plot

```python
import matplotlib.pyplot as plt

data = [20,30,40,50,30,40,70,80]
plt.violinplot(data, showmedians=True)
plt.show()
```

## Subplots

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.plot([1,2,3],[4,5,6])

plt.subplot(1,2,2)
plt.bar([1,2,3],[3,6,9])

plt.show()
```

## Save Plot

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3],[4,5,6])
plt.savefig("01-python/line_plot.png")
plt.show()
```

# Seaborn

Seaborn is a high-level data visualization library built on top of Matplotlib.
It provides better aesthetics and easier statistical plotting.

## Line Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({"Days":[1,2,3,4,5], "Sales":[50,40,60,30,45]})
sns.lineplot(data=data, x="Days", y="Sales")
plt.show()
```

## Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.barplot(data=tips, x="day", y="tip", hue="sex")
plt.show()
```

## Histogram

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

## Scatter Plot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.show()
```

## Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot = tips.pivot_table(values="tip", index="day", aggfunc="mean")
sns.heatmap(pivot, annot=True)
plt.show()
```

## Count Plot

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

## Box Plot

```python
sns.boxplot(data=tips, x="day", y="tip")
plt.show()
```

## Pair Plot

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

## Violin Plot

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

## Categorical Plot

```python
sns.catplot(data=tips, x="day", y="tip", hue="sex", kind="bar")
plt.show()
```

## Style and Color Palette

```python
sns.set_style("whitegrid")
sns.palplot(sns.color_palette("viridis"))
```

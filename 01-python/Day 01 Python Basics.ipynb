**3. Write a program to create area calculator.**

```python
print(" 1: Rectangle \n 2: Square \n 3 : Triangle \n 4 : Circle")

ch = int(input("Enter number from 1 to 4: "))

if ch == 1:
    l = float(input("Length of rectangle: "))
    b = float(input("Breadth of rectangle: "))
    print("Area of rectangle: ", l * b)
    
elif ch == 2:
    side = float(input("Enter the side of square: "))
    print("Area of square: ", side ** 2)

elif ch == 3:
    base = float(input("Enter the base of triangle: "))
    height = float(input("Enter the height of triangle: "))
    print("Area of triangle: ", 0.5 * base * height)

elif ch == 4:
    radius = float(input("Enter the radius of circle: "))
    print("Area of circle: ", 3.14 * radius * radius)

else:
    print("Invalid Choice..Try Again :) ")
    
```

**4. Write a program to check whether the passed letter is a vowel or not.**

```python
letter = input("Enter the letter: ")

if (letter in "aeiou" or letter in "AEIOU"):
    print("Vowel")
else:
    print("Not a vowel")
```

**5. Write a program to check if a number is a single - digit number, 2 - digit number and so on., up to 5 -digits.**

```python
num = int(input("Enter the number upto 5 - digits: "))

if (num >= 0 and num <= 9):
    print("Number is a single digit number")

elif (num >= 10 and num <= 99):
    print("Number is a 2 - digit number")

elif (num >= 100 and num <= 999):
    print("Number is a 3 - digit number")

elif (num >= 1000 and num <= 9999):
    print("Number is a 4 - digit number")

elif (num >= 10000 and num <= 99999):
    print("Number is a 5 - digit number")

else:
    print("Enter the number within the range :( ")
```

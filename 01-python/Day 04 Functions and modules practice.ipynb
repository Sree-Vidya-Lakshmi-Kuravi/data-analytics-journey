
**Function to calculate sum and average**


```python
def calc_sum_and_avg(numbers):
    total_sum = sum(numbers)
    avg = total_sum / len(numbers)
    return total_sum, avg

numbers = [10, 39, 45, 12]
print(calc_sum_and_avg(numbers))
```



**Function to check even or odd**


```python
def check_even_or_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

print(check_even_or_odd(15))
print(check_even_or_odd(26))
```


**Function to calculate area (square / rectangle)**


```python
def area_of_square(side):
    return "Area of Square:", side**2

def area_of_rect(l, b):
    return "Area of Rectangle:", l*b

print(area_of_square(12))
print(area_of_rect(32, 5))
```


**One lambda function**


```python
y = lambda x: x * x
print(y(12))
```


**Use random module to:**

**-- Generate a random number**

**-- Pick a random element from a list**


```python
import random
x = random.randint(3, 89)
print("Random number:", x)

y = random.choice(["Siri", "Sree", "Vidya", "Lakshmi"])
print("Random element from list:", y)
```

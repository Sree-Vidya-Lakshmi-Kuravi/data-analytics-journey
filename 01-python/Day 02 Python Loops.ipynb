**Loops**

**for loop**

```python
for i in range(1, 4):
    print(i, end = " ")
```

```python
for i in range(1, 5, 2):
    print(i, end = " ")
```

```python
for i in range(6):
    print(i, end = " ")
```

**while loop**

```python
n = 3

while n<=6:
    print(n, end = " ")
    n += 1
```

```python
#Multiplication table using while loop

n = 1
num = int(input("Enter the numnber you want as a table: "))

while n <= 10:
    print(num, "x", n, "=", n*num)
    n += 1
```

**while True**

```python
while True:
    a = int(input("Num 1: "))
    b = int(input("Num 2: "))
    print("Sum: ", a + b)

    repeat = input("Do you want to stop here?")
    if repeat == "yes":
        break
```

**Nested loops**

```python
for i in range(1,4):
    for j in range(1,11):
        print(j, end = " ")
    print()
```

```python
for i in range(1, 6): #rows
    for j in range(1, i+1):  #columns should increase
        print(j, end = " ")
    print()
```

**for loop with conditional statments**

```python
for i in range(1, 11):
    if i == 3:
        print("Favourite")
    else:
        print(i, end = " ")
```

```python
for i in range(1, 46):
    if (i % 8 ==0 and i % 12 == 0):
        print(i, end = " ")
```

```python
i = 1
while i<=10:
    if i%3 == 1 :
        print(i, end = " ")
    i += 1
```

**break statement**

```python
for i in range(1, 11):
    if i==7:
        break
    else:
        print(i, end = " ")
```

**continue statement**

```python
for i in range(1, 11):
    if i==7:
        continue
    else:
        print(i, end = " ")
```

**PROBLEM SOLVING**

**01. Write a program to find a sum of all the even numbers upto 50.**

```python
sum = 0
for i in range(1, 51):
    if (i%2 ==0):
        sum+=i
print(sum)
    
```

**02. Write a program to write first 20 numbers and their squared numbers.**

```python
for i in range(1, 21):
    print("Number: ", i, "Squared number: ", i**2)
```

**03. Write a program to find sum of first 10 odd numbers using while loop.**

```python
num = 1
sum = 0

while num<=20:
    if num%2 != 0:
        sum += num
    num += 1
print(sum)
    
```

**04. Write a program to check if a number is divisible by 8 and 12 upto 100 numbers.**

```python
for i in range(1, 101):
    if (i%8 == 0 and i%12 == 0):
        print(i, end = " ")
    
```

**05. Write a program to create a billing system at supermarket.**

```python
print("Welcome to supermarket billing..")

while True:
    name = input("Enter customer's name: ")
    total = 0

    while True:
        print("Enter the amount and quantity: ")
        amount = float(input("Enter amount: "))
        qty = float(input("Enter quantity: "))
        total += amount*qty

        repeat = input("Add more items? (y/n) ")
        if repeat == 'n' or repeat == 'N':
            break
    print("=" * 30)
    print("Name: ", name)
    print("Amount to be paid: ", amount)
    print("Thank You")

    rep1 = input("Do you want to go to next customer? (y/n) ")
    if rep1 == "n" or rep1=="N":
        break
```

**Pattern Loops**

```python
for i in range(1, 6):
    for j in range(5, i, -1):   #starts with 4 and move backwards till 1
        print(" ", end = " ")
    for k in range(i):
        print("*", end = " ")
    print()
```

```python
for i in range(1, 6):
    for j in range(i, 0, -1):
        print(j, end = " ")
    print()
```

```python
for i in range(1,6):
    for j in range(0,6-i):
        print(i,end=" ")
    print()
```

```python
for i in range(1,6):
    print("  "*(5-i),"* "*i)
```

```python
n = 5

for i in range(1, 2*n):
    if i <= n:
        print("* " * i)
    else:
        print("* " * (2*n - i))
```

```python
for i in range(1,9):
    for j in range(1,i+1):
        print(i*j,end=" ")
    print()
```

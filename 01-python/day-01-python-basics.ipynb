## ðŸ”¹ Problem 1: Sum and Count

### Problem Statement
Write a function that returns:
- Total sum of numbers
- Count of numbers

### Approach
Use built-in Python functions:
- sum()
- len()

```python
#Solution
#Function to return mean of numbers
numbers=[1,2,3,4,5] 
def average(numbers): 
    return sum(numbers)/len(numbers) 
print(average(numbers))
```

## ðŸ”¹ Problem 2: Max and Min Finder

### Problem Statement
Find the maximum and minimum values from a list.

### Approach
Use Python built-in functions:
- max()
- min()

```python
### Solution
#Function to find min and max
def min_max_finder(num):
    minimum = min(num)
    maximum = max(num)
    print(minimum, maximum)
    return minimum, maximum

num = [2,6,8,4,1]
print(min_max_finder(num))
```

## ðŸ”¹ Problem 3: Average of Even Numbers Only

### Problem Statement
Calculate the average of only even numbers from a list.

### Approach
- Filter even numbers using modulo
- Compute average using sum and len

```python
#Solution
#Average of Even Numbers Only

def avg_of_even(num):
   for n in num:
       if n%2 == 0:
           return sum(num)/len(num)

numbers = [2, 5, 4, 7, 8]
print(avg_of_even(numbers))
```

## ðŸ”¹ Problem 4: Count Numbers Above Average

### Problem Statement
Calculate the average and count how many numbers are greater than the average.

### Why This Matters
This logic is frequently used in data analysis to identify:
- High-performing customers
- Above-average sales
- Top-performing products

```python
#Count Numbers Above Average

def num_above_average(num):
    avg = sum(num)/len(num)
    count = 0

    for n in num:
        if n>avg:
            count += 1

    return n, count, avg

num = [4,9,3,1]
print(num_above_average(num))
```

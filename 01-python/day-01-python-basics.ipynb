```python
#Average of Even Numbers Only

def avg_of_even(num):
   for n in num:
       if n%2 == 0:
           return sum(num)/len(num)

numbers = [2, 5, 4, 7, 8]
print(avg_of_even(numbers))
```

```python
#Function to find min and max
def min_max_finder(num):
    minimum = min(num)
    maximum = max(num)
    print(minimum, maximum)
    return minimum, maximum

num = [2,6,8,4,1]
print(min_max_finder(num))
```

```python
#Count Numbers Above Average

def num_above_average(num):
    avg = sum(num)/len(num)
    count = 0

    for n in num:
        if n>avg:
            count += 1

    return n, count, avg

num = [4,9,3,1]
print(num_above_average(num))
```

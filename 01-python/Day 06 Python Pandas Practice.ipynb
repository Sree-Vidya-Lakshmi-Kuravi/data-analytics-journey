
**Filling the missing values with mean and median**


```python
print("FILLING THE MISSING VALUES WITH MEAN:\n")
amount_avg = df["Amount"].mean()
df["Amount"] = df["Amount"].fillna(amount_avg)
print(df)
print()
print("----------------------------------------------------------------------------------------------------------------")
print("FILLING THE MISSING VALUES WITH MEDIAN:\n")
amount_median = df["Amount"].median()
df["Amount"] = df["Amount"].fillna(amount_median)
print(df)
```


**GROUP BY**


```python
df = pd.read_excel("E:/Vidya Career/IT JOB/Data Analyst Path/Excel/Pandas practice.xlsx")
print("DataFrame:\n\n",df)
```

```python
group_by_data = df.groupby("Name")["Amount"].mean()
print(group_by_data)
```


**PIVOT**


```python
#rows becomes columns
pivot_val =df.pivot(index = "S.No", columns = "Street Food", values = "Amount")
print(pivot_val)
```


**MELT**


```python
#columns becomes rows
melt_val = df.melt(id_vars = "S.No", var_name = "col_names", value_name = "row_names")
print(melt_val)
```


**MERGE**


```python
students = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Asha", "Rahul", "Sneha"]
})

marks = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Marks": [88, 75, 95]
})

merge_val = pd.merge(students, marks, on = "StudentID", how = "inner")
print(merge_val)
```


**Key Takeaways**

  -- Pivot - rows becomes columns

  -- Melt - columns becomes rows

  -- Merge - combines the values

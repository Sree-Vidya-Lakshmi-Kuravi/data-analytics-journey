**NumPy Arrays**


***Import the module***


```python
import numpy as np
```


***Create 1D and 2D arrays***


```python
import numpy as np

#1-D array
a = np.array([23, 45, 21, 53])
print("1D array:", a)

#2-D array
b = np.array([[32, 43, 54, 65], [12, 34, 56, 78]])
print("2D array:\n", b)
```


***Check shape, size, datatype***


```python
import numpy as np

b = np.array([[32, 43, 54, 65], [12, 34, 56, 78]])
print("Shape:", np.shape(b))
print("Size:", np.size(b))
print("Datatype:", np.dtype)
```


***Indexing & slicing***


```python
#indexing
print(a[1])
print(b[1])
print(b[1][2])

#slicing
print(a[0:2])
print(b[::])
print()
print(b[:, 1])
```


**Operations**

***Addition, subtraction, multiplication***


```python
import numpy as np

a = np.array([2, 4, 6, 9, 10])
b = np.array([1, 3, 5, 7, 8])
print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
```


***Aggregate Functions***


```python
import numpy as np

a = np.array([2, 4, 6, 9, 10])
print("Sum:", np.sum(a))
print("Minimum:", np.min(a))
print("Maximum:", np.max(a))
print("Mean:", np.mean(a))
```
